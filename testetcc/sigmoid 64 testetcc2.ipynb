```python
# Imports
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
# Inicializando a Rede Neural Convolucional
# print(help(keras.layers.Conv2D))
classifier = Sequential()
classifier.add(Conv2D(16, (3,3), input_shape = (128, 128, 3), activation = 'relu'))
```

```python
# Passo 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
# Adicionando a Segunda Camada de Convolução
classifier.add(Conv2D(16, (3,3), activation = 'relu'))
```

```python
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
# Passo 3 - Flattening
classifier.add(Flatten())
```

```python
# Passo 4 - Full connection
classifier.add(Dense(units = 256, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# Compilando a rede
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy', 'mse'])
```

```python
# Criando os objetos train_datagen e validation_datagen com as regras de pré-processamento das imagens
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                    # shear_range = 0.2,
                                    # zoom_range = 0.2,
                                   horizontal_flip = True)

validation_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Pré-processamento das imagens de treino e validação
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/tentativa3/treino',
                                                 target_size = (128, 128),
                                                 batch_size = 30,
                                                 #color_mode='grayscale',
                                                 class_mode = 'binary')

validation_set = validation_datagen.flow_from_directory('/content/drive/MyDrive/tentativa3/validacao',
                                                        target_size = (128, 128),
                                                        batch_size = 30,
                                                        #color_mode='grayscale',
                                                        class_mode = 'binary')
```

```python
classifier.fit(training_set, epochs = 100, validation_data = validation_set, validation_steps = 30)
```
