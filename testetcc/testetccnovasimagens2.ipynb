```python
VP = [16, 15, 14, 12, 13, 13, 10, 13, 15, 13, 14, 13]
FP = [6, 11, 6, 5, 9, 6, 2, 8, 7, 6, 4, 12]
FN = [2, 3, 4, 6, 5, 5, 8, 5, 3, 5, 4, 5]
VN = [17, 12, 17, 18, 14, 17, 21, 15, 16, 17, 19, 11]

if (len(VP) == 12 and len(FP) == 12 and len(FN) == 12 and len(VN) == 12):
  print("tudo certo")
else:
  print("erro")

preh = []
prep = []
senh = []
senp = []
for x in range(0 , len(VP)):
  resultado =  VP[x]/(VP[x]+FP[x])
  preh.append(round(resultado,2))
print(f'pre h : {preh}')
for x in range(0 , len(VP)):
  resultado =  VN[x]/(VN[x]+FN[x])
  prep.append(round(resultado,2))
print(f'pre p : {prep}')
for x in range(0 , len(VP)):
  resultado =  VP[x]/(VP[x]+FN[x])
  senh.append(round(resultado,2))
print(f'sen h : {senh}')
for x in range(0 , len(VP)):
  resultado =  VN[x]/(VN[x]+FP[x])
  senp.append(round(resultado,2))
print(f'sen p : {senp}')
f1h = []
f1p = []
for x in range(0 , len(VP)):
  resultado =  (2*preh[x]*senh[x])/(preh[x]+senh[x])
  f1h.append(round(resultado,2))
print(f'f1 h : {f1h}')
for x in range(0 , len(VP)):
  resultado =  (2*prep[x]*senp[x])/(prep[x]+senp[x])
  f1p.append(round(resultado,2))
print(f'f1 p : {f1p}')
```

```python
# Imports
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
# Inicializando a Rede Neural Convolucional
# print(help(keras.layers.Conv2D))
classifier = Sequential()
classifier.add(Conv2D(32, (3,3), input_shape = (128, 128, 3), activation = 'relu'))
```

```python
# Passo 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
# Adicionando a Segunda Camada de Convolução
classifier.add(Conv2D(32, (3,3), activation = 'relu'))
```

```python
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
# Passo 3 - Flattening
classifier.add(Flatten())
```

```python
# Passo 4 - Full connection
classifier.add(Dense(units = 256, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# Compilando a rede
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy', 'mse'])
```

```python
# Criando os objetos train_datagen e validation_datagen com as regras de pré-processamento das imagens
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                    # shear_range = 0.2,
                                    # zoom_range = 0.2,
                                   horizontal_flip = True)

validation_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Pré-processamento das imagens de treino e validação
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/tentativa3/treino',
                                                 target_size = (128, 128),
                                                 batch_size = 30,
                                                 #color_mode='grayscale',
                                                 class_mode = 'binary')

validation_set = validation_datagen.flow_from_directory('/content/drive/MyDrive/tentativa3/validacao',
                                                        target_size = (128, 128),
                                                        batch_size = 30,
                                                        #color_mode='grayscale',
                                                        class_mode = 'binary')
```

```python
classifier.fit(training_set, epochs = 100, validation_data = validation_set, validation_steps = 30)
```
